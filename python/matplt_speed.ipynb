```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload your CSV file (replace 'your_file.csv' with the correct path if using locally)
file_path = '/content/LoRa Binghamton - HELTEC LoRa 32 - mapping_GPS_track.csv'  # Update with your file path
data = pd.read_csv(file_path)

# Convert Time column to datetime and sort data
data['Time'] = pd.to_datetime(data['Time'])
data = data.sort_values(by='Time')

# Display the first few rows of data
print("Dataset preview:")
print(data.head())

# Correlation analysis between Speed, RSSI, and SNR
correlations = data[['Speed (mph)', 'RSSI', 'SNR']].corr()
print("Correlation analysis:")
print(correlations)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the Airtime column is numeric
data['Airtime (s)'] = data['Airtime'].str.rstrip('s').astype(float)  # Remove 's' and convert to float

# Scatterplot for Airtime vs RSSI and Airtime vs SNR
plt.figure(figsize=(12, 6))

# Airtime vs RSSI
plt.subplot(1, 2, 1)
sns.scatterplot(x=data['Airtime (s)'], y=data['RSSI'], hue=data['RSSI'], palette='coolwarm', s=50)
plt.title('Airtime vs RSSI', fontsize=14)
plt.xlabel('Airtime (s)', fontsize=12)
plt.ylabel('RSSI', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Airtime vs SNR
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['Airtime (s)'], y=data['SNR'], hue=data['SNR'], palette='viridis', s=50)
plt.title('Airtime vs SNR', fontsize=14)
plt.xlabel('Airtime (s)', fontsize=12)
plt.ylabel('SNR', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Scatterplots for RSSI and SNR vs Speed
plt.figure(figsize=(12, 6))

# RSSI vs Speed
plt.subplot(1, 2, 1)
sns.scatterplot(x=data['Speed (mph)'], y=data['RSSI'], hue=data['RSSI'], palette='viridis', s=50)
plt.title('RSSI vs Speed', fontsize=14)
plt.xlabel('Speed (mph)', fontsize=12)
plt.ylabel('RSSI', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# SNR vs Speed
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['Speed (mph)'], y=data['SNR'], hue=data['SNR'], palette='coolwarm', s=50)
plt.title('SNR vs Speed', fontsize=14)
plt.xlabel('Speed (mph)', fontsize=12)
plt.ylabel('SNR', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Scatterplots for RSSI and SNR vs Speed
plt.figure(figsize=(12, 6))

# RSSI vs Speed
plt.subplot(1, 2, 1)
sns.scatterplot(x=data['SF'], y=data['RSSI'], hue=data['RSSI'], palette='viridis', s=30, marker='x')
plt.title('SF vs RSSI', fontsize=14)
plt.xlabel('Spreading Factor (SF)', fontsize=12)
plt.ylabel('RSSI', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# SNR vs Speed
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['SF'], y=data['SNR'], hue=data['SNR'], palette='viridis', s=30, marker='x')
plt.title('SF vs SNR', fontsize=14)
plt.xlabel('Spreading Factor (SF)', fontsize=12)
plt.ylabel('SNR', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
